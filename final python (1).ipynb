```python
status_count=df['Booking_Status'].value_counts()
plt.figure(figsize=(8,4))
plt.pie(status_count,labels=status_count.index,autopct='%1.1f%%')
plt.title("Ride status distribution")
plt.axis("equal")
plt.show()
```

```python
x=df.groupby('Payment_Method')['Booking_Value'].sum()
plt.figure(figsize=(8,4))
plt.bar(x.index.astype(str),x.values,color="red")
plt.title("revenue by payment type")
plt.xlabel("payment method")
plt.ylabel("revenue")
plt.show()
```

```python
top_location=df['Pickup_Location'].value_counts().head(5)
plt.figure(figsize=(8,5))
top_location.plot(kind='bar',color='green')
plt.title("top booking location")
plt.xlabel("pickup location")
plt.ylabel("ride count")
plt.show()
```

```python
df['date']=pd.to_datetime(df['Date'],format='%d-%m-%Y',errors='coerce')
month=df.groupby(df['date'].dt.to_period('M'))['Booking_Value'].sum().reset_index()
month['date']=month['date'].astype(str)
plt.figure(figsize=(8,4))
plt.plot(month['date'],month['Booking_Value'], marker='o', color='pink')
plt.title("Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[["Booking_Value","Ride_Distance","Customer_Rating","Driver_Ratings"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
